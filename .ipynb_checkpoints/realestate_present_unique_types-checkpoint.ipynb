# Real Estate Report

## show real estate types

```python
import pandas as pd

# import realstate data from csv spplited by ;
realestate_data = pd.read_csv("data/aluguel.csv", sep=';')
```

```python
realestate_data.head()
```

```python
real_estate_types = realestate_data["Tipo"]
```

```python
type(real_estate_types)
```

```python
# remove duplicates
real_estate_types.drop_duplicates(inplace=True)
```

```python
# sort ascending
real_estate_types.sort_values(ascending=True)
```

## presenting the data

```python
# create a dataframe
real_estate_types = pd.DataFrame(real_estate_types)
real_estate_types
```

```python
# set a new index
real_estate_types.index = range(len(real_estate_types.index))
# set a name for index column
real_estate_types.columns.name = "Name"
# show it
real_estate_types
```
