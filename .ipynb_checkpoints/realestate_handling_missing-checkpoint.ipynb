## removing nan from our dataframe

```python
import pandas as pd
df_data = pd.read_csv('data/aluguel.csv', sep=';')
df_data.head(10)
```

```python
# identify null values
df_data.isnull()
```

```python
# identify null values consolidate by columns
df_data.info()
```

```python
# bring only records that has VALOR null
df_data[df_data['Valor'].isnull()]
```

```python
# remove NaN records
size_df = df_data.shape[0]
size_df
```

```python
df_data.dropna(subset = ['Valor'], inplace=True)
size_dt_without_nan = df_data.shape[0]
size_dt_without_nan
```

```python
print("{} dropped records".format(size_df - size_dt_without_nan))
```

## handling nan in specific cases

```python
df_data.head()
```

```python
# identify null condominio
df_data[df_data['Condominio'].isnull()]
```

```python
# Task - remove only Apartments that have nan condominios
select_for_delete = (df_data['Tipo'] == 'Apartamento') & (df_data['Condominio'].isnull())
rows_before_delete = df_data.shape[0]
df_data = df_data[~select_for_delete] # the ~ sinal change True to False and vice versa
rows_after_delete =  df_data.shape[0]
print("Apartments deleted : {}".format(rows_before_delete - rows_after_delete))
```

```python
df_data[df_data['Condominio'].isnull()].shape[0]
```

```python
# fill NaN with zero
df_data_2 = df_data.fillna({"Condominio" : 0, "IPTU": 0})
df_data_2.head(10)
```
