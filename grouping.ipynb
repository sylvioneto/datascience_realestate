# Grouping

```python
import pandas as pd
```

```python
df_rent = pd.read_csv('data/aluguel.csv', sep=';')
df_rent.head(10)
```

```python
# Computing mean by location

# location list with only the wanted locations
location = ['Copacabana', 'Jardim Botânico', 'Barra da Tijuca', 'Centro', 'Higienópolis']
selection = df_rent['Location'].isin(location)

# create a df with only those locations records
df_location = df_rent[selection]

# create a df grouped by location
location_group = df_location.groupby('Location')
location_group
```

```python
# for each group, print the mean
for loc, data in location_group:
    print("{} -> {}".format(loc, data['Rent'].mean()))
```

```python
# without print function
location_group['Rent'].mean()
```

```python
# with two columns
location_group['Rent', 'Condo fee'].mean().round(2)
```

# Descriptive statistics

```python
# get descriptive statistics
location_group['Rent'].describe().round(2)
```

```python
# get only specific columns
location_group['Rent'].aggregate(['min', 'max', 'sum'])
```

```python
# using matplot to create a graphic
# setup matplot
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = location_group['Rent'].std().plot.bar(color = 'blue')
fig.set_ylabel('Rent price')
fig.set_title('Averate Rent price'), {'fontsize': 22}
```
