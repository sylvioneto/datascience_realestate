# Verify for missing data

```python
import pandas as pd
df_data = pd.read_csv('data/aluguel.csv', sep=';')
df_data.head(10)
```

```python
# identify null values
df_data.isnull()
```

```python
# identify null values consolidate by columns
df_data.info()
```

```python
# bring only records that has VALOR null
df_data[df_data['Valor'].isnull()]
```

```python
# remove NaN records
size_df = df_data.shape[0]
size_df
```

```python
df_data.dropna(subset = ['Valor'], inplace=True)
size_dt_without_nan = df_data.shape[0]
size_dt_without_nan
```

```python
print("{} dropped records".format(size_df - size_dt_without_nan))
```
